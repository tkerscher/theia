# Scene Render

Since it's hard to debug images, we'll instead have this notebook to see if the
`SceneRender` works

```python
import numpy as np
import matplotlib.pyplot as plt
from theia.scene import MeshStore, Scene, SceneRender, Transform
```

```python
# load meshes
store = MeshStore({
    "cone" : "../assets/cone.stl",
    "cube" : "../assets/cube.stl",
    "monkey" : "../assets/suzanne.stl",
    "sphere" : "../assets/sphere.stl",
    "torus" : "../assets/torus.stl",
})
```

```python
# create some instances
torus = store.createInstance("torus")
torus.transform = Transform.Translation(-2.0,0.0,0.0).numpy()
cone1 = store.createInstance("cone")
cone1.transform = Transform.Translation(2.0,0.5,-1.0).numpy()
cone2 = store.createInstance("cone")
cone2.transform = Transform.Translation(0.5,0.5,-1.0).scale(0.8,0.8,1.5).numpy()
monkey = store.createInstance("monkey")
monkey.transform = Transform.Rotation(1.0,0.0,1.0,-0.8).translate(0.0,-0.5,0.0).scale(0.8,0.8,0.8).numpy()
```

```python
# create scene
scene = Scene([torus, cone1, monkey, cone2], {})
```

```python
# create renderer
renderer = SceneRender(
    dimension=(5.0,5.0),
    position=(0.0,-2.0,0.0),
    direction=(0.0,1.0,0.0))
```

```python
# render scene
image = renderer.render(scene, position=(-1.0,-2.0,0.0), direction=(0.0,1.0,-0.2))

plt.figure(figsize=(8,8))
plt.imshow(image)
plt.show()
```

```python
# render scene from different angle
image = renderer.render(scene, position=(-2.0,-2.0,1.0), direction=(0.8,1.0,-0.3))

plt.figure(figsize=(8,8))
plt.imshow(image)
plt.show()
```
